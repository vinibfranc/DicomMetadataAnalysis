# Intro

## Visão geral  

O conjunto de dados é projetado para permitir que diferentes métodos sejam testados para examinar as tendências nos dados de imagem de TC associados ao uso de contraste e idade do paciente. A idéia básica é identificar texturas de imagem, padrões estatísticos e características que se correlacionam fortemente com essas características e possivelmente construir ferramentas simples para classificar automaticamente essas imagens quando tiverem sido classificadas erroneamente (ou encontrar outliers que podem ser casos suspeitos, medições ruins ou máquinas mal calibradas).

## Dados
O conjunto de dados consiste em 50 imagens de câncer de pulmão. Mais detalhes no relatório! <br> Link do dataset completo - https://wiki.cancerimagingarchive.net/display/Public/LIDC-IDRI

### Processando dados DICOM

Processando os arquivos DICOM com base no arquivo csv de metadados anteriormente lido

```python
def process_data(path):
    dicom_data = []
    for filepath in glob(PATH+path):
        dicom_features = dicom.read_file(filepath)
        dicom_data.append({'path': filepath, 
                     'ID':dicom_features.PatientID,
                     'Modality':dicom_features.Modality,
                     'Manufacturer': dicom_features.Manufacturer})
    result = pd.DataFrame(dicom_data)
    result.head()
    return result
```

```python
dicom_data = process_data('dataset/*.dcm')
```

# Explorando dados

## Verificando consistência entre csv e imagens

Comparando as ocorrências de determinadas características dentro do conjunto de dados

```python
def countplot_comparison(feature):
    fig, (ax1, ax3) = plt.subplots(1,2, figsize = (16, 4), squeeze=True)
    s1 = sns.countplot(data_df[feature], ax=ax1)
    s1.set_title("Overview data")
    s3 = sns.countplot(dicom_data[feature], ax=ax3)
    s3.set_title("Dicom files data")
    plt.show()
```

```python
# Mostra a distribuição de modalidade de imagens no dataset (cerca de 1000 são de CT e 200 de DX por exemplo)
# Todas as 50 imagens escolhidas são de CT

countplot_comparison('Modality')
```

```python
# Todas as 50 imagens escolhidas são do fornecedor GE MEDICAL SYSTEMS

countplot_comparison('Manufacturer')
```

## Mostrando imagens DICOM e metadados iniciais

```python
def show_images(data):
    img_data = list(data[:50].T.to_dict().values())
    f, ax = plt.subplots(10,5, figsize=(16, 20))
    for i,data_row in enumerate(img_data):
        data_row_img = dicom.read_file(data_row['path'])
        
        modality = data_row_img.Modality
        patient = data_row_img.PatientID
        manufacturer = data_row_img.Manufacturer
        
        ax[i//5, i%5].imshow(data_row_img.pixel_array, cmap=plt.cm.bone) 
        ax[i//5, i%5].axis('off')
        ax[i//5, i%5].set_title('Patient: {}\nModality: {}\n{}'.format(patient, modality, manufacturer))
        plt.subplots_adjust(hspace=0.8, wspace=0)
    plt.show()
```

```python
show_images(dicom_data)
```

Um arquivo DICOM contém muito mais informações do que a imagem que representamos. Vamos dar uma olhada, para um dos arquivos DICOM, esta informação. Vamos ler apenas o primeiro arquivo dicom e mostrar esta informação.

```python
dicom_file_path = list(dicom_data[:1].T.to_dict().values())[0]['path']
dicom_file_dataset = dicom.read_file(dicom_file_path)
dicom_file_dataset
```

### Escolha das características a serem analisadas

Foram escolhidas as seguintes características, pois me pareceram relavantes para a análise das imagens:

* ImageType
* Contrast/Bolus Agent
* Body Part Examined
* Exposure Time
* Filter Type
* Convolution Kernel
* Slice Location

```python
def process_new_features(dicom_data, path):
    
    dicom_data = []
    for filepath in glob(PATH+path):
        data_row_img = dicom.read_file(filepath)
        dicom_data.append({'path': filepath, 
                     'ID':data_row_img.PatientID,
                     'Modality': data_row_img.Modality,
                     'BodyPart': data_row_img.BodyPartExamined,
                     'ConvKernel': data_row_img.ConvolutionKernel,
                     'SliceLocation': data_row_img.SliceLocation,
                     'Filter': data_row_img.FilterType,
                     'Exposure': data_row_img.ExposureTime,
                     })
    result = pd.DataFrame(dicom_data)
    result.head()
    return result
```

```python
dicom_new_data = process_new_features(dicom_data, 'dataset/*.dcm')
```

## Mostrando imagens com alguns metadados extraídos anteriormente

```python
def show_dicom_images(data):
    img_data = list(data[:50].T.to_dict().values())
    f, ax = plt.subplots(10,5, figsize=(16,20))
    for i,data_row in enumerate(img_data):

        data_row_img = dicom.read_file(data_row['path'])
        modality = data_row_img.Modality
        slice_location = data_row_img.SliceLocation
        body_part = data_row_img.BodyPartExamined
        
        ax[i//5, i%5].imshow(data_row_img.pixel_array, cmap=plt.cm.bone) 
        ax[i//5, i%5].axis('off')
        ax[i//5, i%5].set_title('Patient: {}\nModality: {}\nSlice Location: {}\nBody Part: {}'.format(
         data_row['ID'], modality, slice_location, body_part))
        plt.subplots_adjust(hspace=1.2, wspace=0)
    plt.show()
```

```python
show_dicom_images(dicom_data)
```
